```python
import os
import pandas as pd

print(os.listdir("Datasets/Data_Experiment_1/"))  # Проверим файлы в папке

df = pd.read_excel("Datasets/Data_Experiment_1/pi2.xlsx", engine="openpyxl")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["Humidity"] = pd.to_numeric(df["Humidity"], errors="coerce")
df["Temperature"] = pd.to_numeric(df["Temperature"], errors="coerce")

df["time"] = pd.to_datetime(df["time"], errors="coerce")

df.info()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df[["Humidity", "Temperature"]].boxplot()
plt.title("Boxplot of Humidity and Temperature")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["time"], df["Temperature"], marker="o", linestyle="-")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Temperature Over Time")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
df.isnull().sum()
```

```python
df.fillna(method="ffill", inplace=True)  # Заполнит пропущенные значения предыдущим
```

```python
df.ffill(inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df["time"], df["Temperature"], marker="o", linestyle="-", label="Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Change Temperature with time")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["time"], df["Humidity"], marker="o", linestyle="-", label="Humidity", color="green")
plt.xlabel("Time")
plt.ylabel("Humidity")
plt.title("Change Humidity with time")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
